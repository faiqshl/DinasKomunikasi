```python
import pandas as pd
import numpy as np
import statistics
from statistics import stdev
from datetime import datetime
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_excel('Data Posbindu 1.xlsx')
df
```

```python
now = datetime.now()
df['tahun_lahir'] = pd.DatetimeIndex(df['tanggal_lahir']).year
for index, row in df.iterrows():
    df.loc[index, 'umur'] = now.year - int(row['tahun_lahir'])
```

```python
data = df[['peserta_id','jenis_kelamin','tanggal_lahir','umur','nama_opd',
           'ptm_keluarga','merokok','konsumsi_alkohol','kurang_aktivitas_fisik',
           'sistol','diastol','tanggal']]
data
```

```python
data['tanggal'] = pd.to_datetime(data['tanggal'])
data['bulan'] = pd.DatetimeIndex(data['tanggal']).month
data
```

```python
data = data.dropna()
data
```

```python
data.duplicated().any()
```

```python
data = pd.DataFrame(data)
data.convert_dtypes(infer_objects=True, convert_string=True)
data = data.convert_dtypes()
```

```python
data = data[data.umur != 0]
data = data[data.umur != 1]
data = data[data.umur != 8]
data.sort_values(by='umur')
```

```python
for index, row in data.iterrows():
    
    #cowo
    if row['jenis_kelamin'] == 'laki-laki':
        if row['umur'] > 60:
            if row['sistol'] > 163 & row['diastol'] > 86:
                    data.loc[index, 's_tekanan_darah'] = 'hipertensi'
            elif row['sistol'] > 143 & row['diastol'] > 76:
                    data.loc[index, 's_tekanan_darah'] = 'sedang'
            elif row['sistol'] > 123 & row['diastol'] > 56:
                    data.loc[index, 's_tekanan_darah'] = 'normal'
        elif row['umur'] > 20:
            if row['sistol'] > 139 & row['diastol'] > 89:
                    data.loc[index, 's_tekanan_darah'] = 'hipertensi'
            elif row['sistol'] > 120 & row['diastol'] > 80:
                    data.loc[index, 's_tekanan_darah'] = 'sedang'
            elif row['sistol'] > 89 & row['diastol'] > 59:
                    data.loc[index, 's_tekanan_darah'] = 'normal'
        
    #cewe
    elif row['jenis_kelamin'] == 'perempuan':
        if row['umur'] > 60:
            if row['sistol'] > 160 & row['diastol'] > 87:
                    data.loc[index, 's_tekanan_darah'] = 'hipertensi'
            elif row['sistol'] > 130 & row['diastol'] > 77:
                    data.loc[index, 's_tekanan_darah'] = 'sedang'
            elif row['sistol'] > 110 & row['diastol'] > 57:
                    data.loc[index, 's_tekanan_darah'] = 'normal'
        elif row['umur'] > 20:
            if row['sistol'] > 134 & row['diastol'] > 81:
                    data.loc[index, 's_tekanan_darah'] = 'hipertensi'
            elif row['sistol'] > 115 & row['diastol'] > 71:
                    data.loc[index, 's_tekanan_darah'] = 'sedang'
            elif row['sistol'] > 84 & row['diastol'] > 51:
                    data.loc[index, 's_tekanan_darah'] = 'normal'
                    
```
