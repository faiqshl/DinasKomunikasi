```python
import pandas as pd
from datetime import datetime
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import preprocessing as pre
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_excel('Data Posbindu 1.xlsx')
```

```python
data = data[['peserta_id','tanggal','nama_opd','jenis_kelamin',
           'benjolan_payudara','gula_berlebih','garam_berlebih','lemak_berlebih',
           'merokok','konsumsi_alkohol','kurang_aktivitas_fisik','kurang_buah_sayur']]
data
```

```python
data = data.dropna()
data.isna().sum()
```

```python
number = LabelEncoder()
data['benjolan_payudara'] = number.fit_transform(data['benjolan_payudara'])
data['gula_berlebih'] = number.fit_transform(data['gula_berlebih'])
data['garam_berlebih'] = number.fit_transform(data['garam_berlebih'])
data['lemak_berlebih'] = number.fit_transform(data['lemak_berlebih'])
data['merokok'] = number.fit_transform(data['merokok'])
data['kurang_buah_sayur'] = number.fit_transform(data['kurang_buah_sayur'])
data['kurang_aktivitas_fisik'] = number.fit_transform(data['kurang_aktivitas_fisik'])
data['konsumsi_alkohol'] = number.fit_transform(data['konsumsi_alkohol'])
```

```python
sum(data['benjolan_payudara'] == 1)
```

```python
data['tanggal'] = pd.to_datetime(data['tanggal'])
data['bulan'] = pd.DatetimeIndex(data['tanggal']).month
data = data[['peserta_id','bulan','nama_opd','jenis_kelamin','benjolan_payudara',
             'gula_berlebih','garam_berlebih','lemak_berlebih',
             'merokok','konsumsi_alkohol','kurang_aktivitas_fisik','kurang_buah_sayur']]
```
